```python
def fun1(lst, n):
    if n in lst:
        return True
    else:
        return False
def fun2(lst, n):
    low = 0
    high = len(lst) - 1
    while low <= high:
        middle = (low + high) // 2
        if lst[middle] == n:
            return True
        elif n < lst[middle]:
            high = middle - 1
        elif n > lst[middle]:
            low = middle + 1
    return False
```

```python
lst = [6, 4, 8, 9, 1, 34, 22]
lst.sort()
n = int(input("Enter number to search in list: "))
print(fun1(lst,n))
print(fun2(lst,n))
```
