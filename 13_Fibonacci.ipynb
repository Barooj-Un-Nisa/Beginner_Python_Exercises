```python
n = int(input("Enter how many numbers from fibbonacci series you want to generate: "))
fib(n)
```

```python
def fib(num):
    a = 1
    b = 1
    print(a, end = ' ')
    print(b, end = ' ')
    for i in range(1, (num-2)+1):
        c = a + b
        a = b
        b = c
        print(c, end=" ")
```
