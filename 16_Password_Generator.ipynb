```python
import random

s = "abcdefghijklmnopqrstuvwxyz0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ!@#$%^&*()"

words = ['sun', 'moon', 'star', 'sky']
choice = input("Enter password type (weak/strong)")

if choice == "weak":
    password = random.choice(words) + str(random.randint(10,99))
else: 
    passlen = 8
    password = "".join(random.sample(s,passlen))
print(password)
```
