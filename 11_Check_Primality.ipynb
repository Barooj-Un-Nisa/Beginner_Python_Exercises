```python
def prime(num):
    count = 0
    if num == 1:
        return "Co-prime"
    for i in range(1, num+1):
        if num % i == 0:
            count += 1
    if count == 2:
        return "prime"
    else:
        return "not prime"
        

num = int(input("Enter a number: "))
res = prime(num)
print(f"Number is {res}")
```
