```python
n = int(input("Enter a number: "))
if n % 2 == 0:
    if n % 4 == 0:
        print("You entered a multiple of 4")
    else: 
        print("You enter an even number")
else:
    print("You entered an odd number")
```

```python
num = int(input("Enter num: "))
check = int(input("Enter check: "))
if num % check == 0:
    print(f"{check} divide evenly into num")
else:
    print(f"{check} does not divide evenly into num")
```
